<h4><b>3. Plot a bar graph of Male Vs Female buyers participated in the sales</b></h4>

<h5><i>Plotting a graph with the values of Buyer Gender along the X-axis and their respective frequency along the Y-axis</i></h5>

```python
plt.figure(figsize = (9,5))
sns.countplot(df['Buyer Gender'])
plt.xlabel("Buyer Gender", fontsize='15')
plt.ylabel("No: of Buyers", fontsize='15')
plt.title("Male Vs Female buyers participated in the sales", fontsize='20')
plt.grid()
plt.show()
```

```python
df['Buyer Gender'].value_counts()
```

<h5><i><b><u>Insights:</u></b></br>
4948 Male Buyers and</br> 5052 Female Buyers Found!</br>
Female buyers are more in number than the Male buyers</i></h5>

<h4><b>4. Find the top 5 Cars based on their Sales Price</b></h4>

<h5><i>Extracting the columns Make, Model, Sale Price from the DataFrame and sorting the data in descending order of their Sale Price</i></h5>

```python
df.loc[:, ['Make','Model', 'Sale Price']].sort_values(['Sale Price'], ascending=False).head(5)
```
